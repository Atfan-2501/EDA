TRANSFORM AND APPLY: BOTH SUPPORT LAMBDA FUNCTIONS AND CAN BE USED WITH GROUP BY 

```python
df1 = df.drop("species", axis = 1)
```

```python
def plus_10(x):
    return x+10

def minus_10(x):
    return x-10
```

```python
df1 = df1.apply(plus_10)
df1
```

```python
df1 = df1.transform(minus_10)
df1
```

DIFFERENCES
1. APPLY SUPPORTS AGGREGATE FUNCTIONS AND ALSO ALLOWS PASSING Multiple COLUMNS AS ARGUMEMTS

```python
df1[["sepal_length"]].apply(lambda x: x.mean())
```

```python
def sepal_ratio(df):
    ratio =float(df["sepal_length"])/float(df["sepal_width"])
    return ratio

def petal_ratio(df):
    ratio =float(df["petal_length"])/float(df["petal_width"])
    return ratio
```

```python
df["sepal_ratio"] = df.apply(sepal_ratio, axis= 1)
df["petal_ratio"] = df.apply(petal_ratio, axis = 1)
```

```python
df
```

TRANSFORM ALLOWS PASSING MULTIPLE FUNCTIONS 

```python
df[["petal_length"]].transform([plus_10, minus_10]) 
```
