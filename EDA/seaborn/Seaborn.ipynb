SEABORN

```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = sn.load_dataset("tips")
```

```python
df.head()
```

A. ANALYSIS OF CONTINUOUS VARIABLES USING SEABORN

1. HEATMAP

```python
df.corr()
```

```python
sn.heatmap(df.corr())
```

2. JOINT PLOT

```python
sn.jointplot(x= "total_bill", y= "tip", data= df, kind = "scatter", hue= "sex")
```

```python
sn.jointplot(x= "total_bill", y= "tip", data= df, kind = "hex")
```

```python
sn.jointplot(x= "total_bill", y= "tip", data= df, kind = "reg")
```

```python
sn.jointplot(x= "total_bill", y= "tip", data= df, kind = "hist")
```

```python
sn.jointplot(x= "total_bill", y= "tip", data= df, kind = "kde")
```

Pairplots

```python
sn.pairplot(df, hue = "sex", diag_kind= "hist", palette = "rainbow")
```

```python
sn.pairplot(df, hue = "smoker", diag_kind= "hist")
```

```python
sn.pairplot(df, hue = "sex", diag_kind= "kde", palette = "plasma")
```

DIST PLOT - DENOTES HOW THE COLUMN IS DISTRIBUTED

```python
sn.distplot(df.tip, bins = 10)
```

B. ANALYSIS OF DISCRETE/CATEGORICAL VARIABLES  USING SEABORN

1. Histogram/ Count plot

```python
sn.countplot(df.smoker, hue = df.sex)
```

Barplot

```python
sn.barplot(x = df.smoker, y = "tip", data = df, hue = "sex")
```

BOX PLOTS

```python
sn.boxplot(x = "sex", y = "tip",data= df, palette = "plasma")
```

```python
sn.boxplot(x = "sex", y = "tip",data= df,hue = "smoker", palette = "rainbow")
```

VIOLIN PLOT

```python
sn.violinplot(y = "sex", x = "tip",data= df, palette = "rainbow")
```

```python
sn.violinplot(y = "sex", x = "tip",data= df,hue = "smoker", palette = "rainbow")
```

```python
fig, axes = plt.subplots(2,2, figsize = (11,9))
sn.boxplot(ax = axes[0,0], x = "sex", y = "tip",data= df, palette = "plasma")
sn.boxplot(ax = axes[0,1], x = "sex", y = "tip",data= df,hue = "smoker", palette = "rainbow")
sn.violinplot(ax = axes[1,0], y = "sex", x = "tip",data= df, palette = "rainbow")
sn.violinplot(ax = axes[1,1], y = "sex", x = "tip",data= df,hue = "smoker", palette = "rainbow")
```
